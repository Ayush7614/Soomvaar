### Question: What all business problems you can derive by exploring the data? 

Based on the interpretation of the task:
1. Which products do the most profit making states buy?
2. Which products do the loss bearing states buy?

```python
def state_data_viewer(states):
    # It plots the turnover generated by different product categories and sub-categories for the list of given states.
    product_data = dataA.groupby(['State'])
    for state in states:
        data = product_data.get_group(state).groupby(['Category'])
        fig, ax = plt.subplots(1, 3, figsize = (28,5))
        fig.suptitle(state, fontsize=14)        
        ax_index = 0
        for cat in ['Furniture', 'Office Supplies', 'Technology']:
            cat_data = data.get_group(cat).groupby(['Sub-Category']).sum()
            sns.barplot(x = cat_data.Profit, y = cat_data.index, ax = ax[ax_index])
            ax[ax_index].set_ylabel(cat)
            ax_index +=1
        fig.show()
```

```python
states = ['California', 'Washington', 'Mississippi', 'Arizona', 'Texas']
state_data_viewer(states)
```
